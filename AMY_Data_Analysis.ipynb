# This is the file using to analysis the AMY concentration at 148 different nodes in order to predict the Alzheimer Disease status of the patient. The Alzheimer Disease status could be divided into five stages: CN(normal), SMC, EMCL, LMCL, AD.

```python
import pandas as pd
```

```python
Amy_Data = pd.read_excel('POST_PET_AMY.xlsx')
```

```python
Amy_Data
```

```python
Node_Data = Amy_Data.iloc[:,13:]
```

```python
Node_Data
```

```python
Y_Data = Amy_Data['DX']
```

```python
Y_Data
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components= 20)

principalComponents = pca.fit_transform(Node_Data)
                                        
principalDf = pd.DataFrame(data = principalComponents)
```

```python
explained_variance = pca.explained_variance_ratio_

sum = 0
seventy_five_percent = []

for i in range(len(explained_variance)):
    sum += explained_variance[i]
    if (sum > 0.75):
        break
        
seventy_five_percent = explained_variance[0:i]

print(seventy_five_percent)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

x=[]
count = 0

for i in range(len(explained_variance)):
    count += 1
    x.append(count)

y = explained_variance
ax.bar(x,y)
plt.show()
plt.savefig('AMY_PCA_Components.eps')
```

```python
from sklearn.cluster import KMeans
```

```python
# calculate distortion for a range of number of cluster
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(Node_Data)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
plt.savefig('Elbow_Test_For_K_Means.eps')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(dict(PC1=principalDf.iloc[:,0], PC2=principalDf.iloc[:,1], color = Y_Data))

sns.lmplot('PC1', 'PC2', data=df, hue='color', fit_reg=False)

plt.show()

plt.savefig('2_Component_Clustering.eps')
```

```python
for i in range(len(df.color)):
    
    if (df.color[i] == 'AD' or df.color[i] == 'LMCI' or df.color[i] == 'EMCI'):
        df.color[i] = 'AD, LMCI, or EMCI'
    else:
        df.color[i] = 'CN or SMC'
    
df.color
```

```python
sns.lmplot('PC1', 'PC2', data=df, hue='color', fit_reg=False)

plt.show()

plt.savefig('AMY_2_Component_Clustering_2_Color.eps')
```
